```python
df=pd.read_csv('/kaggle/input/competitions/spaceship-titanic/train.csv')
test_df=pd.read_csv('/kaggle/input/competitions/spaceship-titanic/test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['CryoSleep']=le.fit_transform(df['CryoSleep'])
test_df['CryoSleep']=le.fit_transform(test_df['CryoSleep'])
```

```python
df['Total_Spending']=df['RoomService']+df['FoodCourt']+df['ShoppingMall']+df['Spa']+df['VRDeck']
test_df['Total_Spending']=test_df['RoomService']+test_df['FoodCourt']+test_df['ShoppingMall']+test_df['Spa']+test_df['VRDeck']
```

```python
df['CryoSleep'] = np.where(df['Total_Spending'] > 0, 0, 1)

spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'] 
df.loc[df['CryoSleep'] == True, spending_cols] = 0   
df['Total_Spending']=df['RoomService']+df['FoodCourt']+df['ShoppingMall']+df['Spa']+df['VRDeck']
```

```python
test_df['CryoSleep'] = np.where(test_df['Total_Spending'] > 0, 0, 1)

spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'] 
test_df.loc[test_df['CryoSleep'] == True, spending_cols] = 0   
test_df['Total_Spending']=test_df['RoomService']+test_df['FoodCourt']+test_df['ShoppingMall']+test_df['Spa']+test_df['VRDeck']
```

```python
df.isnull().mean()*100
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['VIP']=df['VIP'].fillna(df['VIP'].mode())
```

```python
test_df['Age']=test_df['Age'].fillna(test_df['Age'].mean())
test_df['VIP']=test_df['VIP'].fillna(test_df['VIP'].mode())
```

```python
df['Surname'] = df['Name'].str.split().str[-1]
df.drop(columns=['Name'],axis=1,inplace=True)
test_df['Surname'] = test_df['Name'].str.split().str[-1]
test_df.drop(columns=['Name'],axis=1,inplace=True)
df.head()
```

```python
family_planet_map = df.groupby('Surname')['HomePlanet'].first()
df['HomePlanet'] = df['HomePlanet'].fillna(df['Surname'].map(family_planet_map))
df['HomePlanet'] = df['HomePlanet'].fillna(df['HomePlanet'].mode()[0])

family_planet_map = test_df.groupby('Surname')['HomePlanet'].first()
test_df['HomePlanet'] = test_df['HomePlanet'].fillna(test_df['Surname'].map(family_planet_map))
test_df['HomePlanet'] = test_df['HomePlanet'].fillna(test_df['HomePlanet'].mode()[0])
```

```python
df.drop(columns=['Surname'],axis=1,inplace=True)

test_df.drop(columns=['Surname'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df['Destination'].unique())
print(df['HomePlanet'].unique())
```

```python
df['HomePlanet']=le.fit_transform(df['HomePlanet'])
df['Destination']=le.fit_transform(df['Destination'])
df['VIP']=le.fit_transform(df['VIP'])
df['Transported']=le.fit_transform(df['Transported'])

test_df['HomePlanet'] = le.fit_transform(test_df['HomePlanet'])
test_df['Destination'] = le.fit_transform(test_df['Destination'])
test_df['VIP'] = le.fit_transform(test_df['VIP'])

```

```python
df['Cabin'].nunique()
```

```python
df['Cabin'] = df['Cabin'].fillna('U/0/U')
df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = df['Cabin'].str.split('/', expand=True)

test_df['Cabin'] = test_df['Cabin'].fillna('U/0/U')
test_df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = test_df['Cabin'].str.split('/', expand=True)
print("Decks:", df['Cabin_Deck'].unique())
print("Sides:", df['Cabin_Side'].unique())
```

```python
df['Cabin_Side']=le.fit_transform(df['Cabin_Side'])

deck_order = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T']
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[deck_order],handle_unknown='use_encoded_value', unknown_value=-1)
df[['Cabin_Deck']] = oe.fit_transform(df[['Cabin_Deck']])

test_df['Cabin_Side']=le.fit_transform(test_df['Cabin_Side'])
test_df[['Cabin_Deck']] = oe.fit_transform(test_df[['Cabin_Deck']])
```

```python
test_df.head()
```

```python
df.drop(columns=['Cabin'],axis=1,inplace=True)
df['Cabin_Num'] = pd.to_numeric(df['Cabin_Num'], errors='coerce')

test_df.drop(columns=['Cabin'],axis=1,inplace=True)
test_df['Cabin_Num'] = pd.to_numeric(test_df['Cabin_Num'], errors='coerce')
```

```python
df['Group_Id'] = df['PassengerId'].str.split('_').str[0].astype(int)
df['Group_Size_Pos'] = df['PassengerId'].str.split('_').str[1].astype(int)
df["GroupSize"] = df.groupby("Group_Id")["PassengerId"].transform("count")
df["IsAlone"]   = (df["GroupSize"] == 1).astype(int)
df.drop('PassengerId', axis=1, inplace=True)

test_df['Group_Id'] = test_df['PassengerId'].str.split('_').str[0].astype(int)
test_df['Group_Size_Pos'] = test_df['PassengerId'].str.split('_').str[1].astype(int)
test_df["GroupSize"] = test_df.groupby("Group_Id")["PassengerId"].transform("count")
test_df["IsAlone"] = (test_df["GroupSize"] == 1).astype(int)
```

```python
df['is_child']=np.where(df['Age']<13,1,0)
df['is_teen']=np.where((df['Age']>=13) & (df['Age']<19),1,0)
df['is_adult']=np.where((df['Age']>=19) & (df['Age']<60),1,0)
df['is_senior']=np.where(df['Age']>=60,1,0)

test_df['is_child'] = np.where(test_df['Age'] < 13, 1, 0)
test_df['is_teen'] = np.where((test_df['Age'] >= 13) & (test_df['Age'] < 19), 1, 0)
test_df['is_adult'] = np.where((test_df['Age'] >= 19) & (test_df['Age'] < 60), 1, 0)
test_df['is_senior'] = np.where(test_df['Age'] >= 60, 1, 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
x=df.drop('Transported',axis=1)
y=df['Transported']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = XGBClassifier(
    n_estimators=1000,
    learning_rate=0.01,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method='hist', 
    device='gpu',      
    random_state=42
)
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python


# 1. Pehle PassengerId ko kahin save kar lo submission ke liye
test_ids = test_df['PassengerId'].copy()

# 2. Ab test_df se PassengerId aur koi bhi bacha hua 'object' column drop karo
# Hum wahi columns use karenge jo training ke 'x' mein hain
test_df_numeric = test_df[x.columns] 

# 3. Ab predict karo (Ab error nahi aayega)
y_pred = model.predict(test_df_numeric)

# 4. Labels convert karo (True/False ya original labels mein)
final_labels = le.inverse_transform(y_pred)

# Ensure karein ki final_labels 'True'/'False' strings ya boolean hain
final_labels = [bool(i) for i in final_labels]

# 3. DataFrame banayein sahi names ke saath
submission = pd.DataFrame({
    'PassengerId': test_ids, # Original IDs from test.csv
    'Transported': final_labels
})

# 4. Save karein
submission.to_csv('submission.csv', index=False)

print("Check clear: Column names are", submission.columns.tolist())
print("Check clear: First 5 labels are", submission['Transported'].head().tolist())

submission.to_csv('submission.csv', index=False)
print("Submission file ready!")
```
